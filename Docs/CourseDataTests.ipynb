How many classes taught by each professor

```python
%%sql
SELECT Faculty_Name, COUNT(Cat_ID) AS Classes_Taught
FROM Courses
GROUP BY Faculty_Name
ORDER BY Classes_Taught DESC
LIMIT 20;
```

Total students professors have taught

```python
%%sql
SELECT Faculty_Name, SUM(Actual) AS Total_Taught
FROM Courses
WHERE Actual > 10
GROUP BY Faculty_Name
ORDER BY Total_Taught DESC
LIMIT 20;
```

What days are popular for certain classes

```python
%%sql
SELECT DISTINCT Cat_ID AS Class_Type, Day, COUNT(Day) AS Num_Days
FROM Courses
    JOIN Course_Meeting USING (CRN)
WHERE Actual > 10
GROUP BY Day, Class_Type
ORDER BY Num_Days DESC
LIMIT 20;
```

What classes are popular with certain professors

Major flaws with capacity and remaining (BI 170, BI 107, AC 11, etc.)

```python
%%sql
SELECT Cat_ID, Faculty_Name, Remaining
FROM Courses
    JOIN Course_Meeting USING (CRN)
ORDER BY Remaining
LIMIT 20;
```

##How many students did internships for credit

```python
%%sql
SELECT Title, Sum(Actual) AS Total_Students
FROM Courses
    JOIN Catalog USING (Cat_ID)
WHERE Title = "Internship"
LIMIT 20;
```

How many students took AC 11 by term

```python
%%sql
SELECT Cat_ID, Term, Sum(Actual) AS Total_Students
FROM Courses
WHERE Cat_ID = "AC 0011"
GROUP BY Term
ORDER BY Total_Students DESC
LIMIT 20;
```

What class by term has had the most sections

```python
%%sql 
SELECT Cat_ID, Term, COUNT(DISTINCT Section) AS Num_Sections
FROM Courses
GROUP BY Term, Cat_ID
ORDER BY Num_Sections DESC
LIMIT 20;
```

What classes has Huntley taught by semester

```python
%%sql
SELECT Faculty_Name, Term, Cat_ID, Section
FROM Courses
WHERE Faculty_Name LIKE ("%Huntley%")
ORDER BY Term;
```
